```python
import pandas
import plotly.express as px
data = pandas.read_csv("results.csv")

fig = px.scatter_mapbox(data, lat = "Latitude", lon = "Longitude", hover_name = "Date", size = "Size", size_max = 8, hover_data = ["Download", "DownloadBytes", "Upload", "UploadBytes", "Latency", "ServerName", "ConnType"], color = "Download", color_continuous_scale=px.colors.sequential.Rainbow, zoom = 7, height = 750)
fig.update_layout(mapbox_style = "open-street-map")
fig.update_layout(margin = {"r":0,"t":0,"l":0,"b":0})
fig.show()
```
